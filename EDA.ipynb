```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%pylab inline
```

```python
df2014 = pd.read_csv('data/sfpd_incident_2014.csv')
df2013 = pd.read_csv('data/sfpd_incident_2013.csv')
df2012 = pd.read_csv('data/sfpd_incident_2012.csv')
df2011 = pd.read_csv('data/sfpd_incident_2011.csv')
df2010 = pd.read_csv('data/sfpd_incident_2010.csv')
df2009 = pd.read_csv('data/sfpd_incident_2009.csv')
df2008 = pd.read_csv('data/sfpd_incident_2008.csv')
df2007 = pd.read_csv('data/sfpd_incident_2007.csv')
datasets = [df2014, df2013, df2012, df2011, df2010, df2009, df2008, df2007]
```

#Let's do a scatter plot on a few cuts

##Crime by Police District 2011-2014

```python
rows = 4
cols = 2

fig, ax = plt.subplots(rows, cols, figsize=(13,21))

colors = plt.cm.Spectral(np.linspace(0, 1, len(df2014['PdDistrict'].unique())))
pds = df2014['PdDistrict'].unique()

for i, df in enumerate(datasets):
    year = pd.to_datetime(df['Date'].values[0]).year    
    ix = i / cols
    iy = i % cols
    
    for p, col in zip(pds, colors):
        label_indices = df['PdDistrict'] == p
        x = df[label_indices]['X']
        y = df[label_indices]['Y']
        ax[ix,iy].plot(x, y, '.',
                       markerfacecolor=col,
                       markeredgecolor='k',
                       alpha=0.1,
                       label=p
                       )
        ax[ix,iy].set_title('Year '+ str(year))
```

####Here, the color differentiates police districts and the shading indicates crime density  
####It looks like crime is pretty consistent through the years, and we can identify specific streets with a lot of crime: mission, geary, 3rd, 19th avenue. Lets superimpose these onto a streetmap later
